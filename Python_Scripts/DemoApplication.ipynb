```python
def is_prime(n):
    """
    This function returns True if n is prime
    otherwise it returns False
    """
    
    for x in range (2, n//2+1):    #This loop checks if x=[2,...,n/2] divides n
        if (n%x==0):               #if it finds at least one divider returns False
            return False
    
    return True 

def prime_factors(n):
    """
    This function returns a list with all the prime factors of n
    """
    
    #initialize variable i and an empty list F
    i = 2    
    F = []

    #if n is prime, n itself is the only prime factor
    if is_prime(n)==True:  
        F.append(n)
        return F
    
    #the loop checks all the prime numbers that divide n and adds them to the list
    while i <= n//2:
        
        if n%i==0:
            n = n // i  #n//=i
            F.append(i)
        
        else:
            i = i + 1  #i+=1 
    
    F.sort() #F=sort(F)
    
    #if the last n is larger than 1, is a prime factor
    #add to the list
    if n>1:
        F.append(n)
    
    return F

def dict_prime_factors(n):
    """
    this function returns a dictionary with the prime factorization of n
    keys are the basis, prime factors
    values are the exponents
    """
    #call prime_factors, L list with n's prime factors
    L=prime_factors(n)
    
    #S new sorted list with unique elements
    S=list(set(L))
    S.sort()
    
    #initialize empty dictionary
    D={}
    
    #build the dictionary
    #for each elements in S (which has unique elements)
    #count how many elements in L (which ha also duplicates)
    
    for i in range (len(S)):
        C=L.count(S[i])
        D[S[i]]=C  #add a new element in the dictionary D[key]=value
    
    return D

def application():
        
    number=-1
    
    #be sure n is a natural number
    while number<0:
        number=int(input("Insert a number: "))
    
    #call dict_prime_factors and print the dictionary
    print (dict_prime_factors(number))
    
    print("\n1.Calculate another factorization \n2.EXIT the application\n")
    
    choice=int(input("Choose an operation: "))
    
    #end application
    if (choice==2):
        print("Goodbye World")
        return
    
    #keep running
    application()
    

    
    
#launch applicatio, NOT indented, NOT function code    
application()
```
